```python

df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python

final_df =pd.merge(df,df_country,on='Country Code',how='left')
```

```python

final_df.head(10)
```

```python

#to check data types
final_df.dtypes
```

```python

final_df.columns
```

```python

country_name=final_df.Country.value_counts().index
```

```python

country_val=final_df.Country.value_counts().values
```

```python

## Pie chart for top 3 country
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

 Observation: Zomato maximum records or transcaction are from India

## numerical variables

```python

final_df.columns
```

```python

ratting=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python

ratting
```


 ## observation
 1. 4.5 to 4.9 ---> Excellent
 2. 4.0 to 4.4 ---> very good
 3. 3.5 to 3.9 ---> good
 4. 2.5 to 3.4 ---> average
 5. 1.8 to 2.4 ---> poor


```python

ratting.head()
```

```python

import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratting)
```

```python

sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color', data=ratting,palette=['white','red','orange','yellow','green','green'])
```


 ## Observation
1. Not Rated count is very high
2. maximun number if rating are between 2.5 to 3.4


```python
## count plot
sns.countplot(x='Rating color', data=ratting ,palette=['blue','red','orange','yellow','green','green'])
```
